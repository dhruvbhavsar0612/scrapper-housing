```python
import pandas as pd
```

```python
df = pd.read_csv('data_with_link.csv', index_col=[0])
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
last_col = df.columns[-2]
df.drop(columns=last_col, inplace=True)
df.rename(columns={'link':last_col}, inplace=True)
```

```python
df.columns
```

```python
# here the columns have anomalies in translation 
df.rename(columns={
    df.columns[0]:'Sr No',
    df.columns[1]:'Document No',
    df.columns[2]:'Document Type',
    df.columns[3]:'D.N. Office',
    df.columns[5]:'Buyer Name',
    df.columns[6]:'Seller Name',
}, inplace=True)
```

```python
df
```

```python
df['Document Type'].unique()
```

```python
df.replace(to_replace='Amphideviate', value='affidavit', inplace=True)
df.replace(to_replace='Deed of Transfer', value='transfer deed', inplace=True)
df.replace(to_replace='leasedeed',value='lease deed',inplace=True)
df.replace(to_replace='Section 66 - Notice of lie pendency', value='Section 66 - Notice of lease pendency', inplace=True)
```

```python
df.dtypes
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='mixed')
```

```python
# df['Sr No'] = df['Sr No'].astype('int32')
# df['Document No'] = df['Document No'].astype('int32')
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = df[col].astype('int32')

df['Document Type'] = df['Document Type'].str.capitalize()
df['Buyer Name'] = df['Buyer Name'].str.capitalize()
df['Seller Name'] = df['Seller Name'].str.capitalize()
```

```python
df.info()
```

```python
df.to_csv('data_with_link_processed.csv', index=False)
```

```python
def split_names(row):
    names = row.split('   ')  # Split by three spaces
    names = [name.strip('0123456789) ') for name in names]  # Remove numbers, closing parenthesis, and spaces
    return names

df['Buyer Name']= df['Buyer Name'].apply(split_names)
df['Seller Name'] = df['Seller Name'].apply(split_names)
```

```python
df = df.explode('Buyer Name').reset_index(drop=True)
df = df.explode('Seller Name').reset_index(drop=True)
```

```python
def remove_numbers_with_parentheses(text):
    import re
    return re.sub(r'\d+\)', '', text)

df['Other information']= df['Other information'].apply(remove_numbers_with_parentheses)
```

```python
df.info()
```

```python
df['Buyer Name'] = df['Buyer Name'].str.strip()
df['Seller Name'] = df['Seller Name'].str.strip()
df['Other information'] =df['Other information'].str.strip()
```

```python
df.info()
```

```python
import numpy as np
# Replace blank spaces with NaN in the entire DataFrame
df.replace('', np.nan, inplace=True)

# Now you can use dropna() to remove rows with NaN values
df.dropna(inplace=True)

# Reset the DataFrame index if needed
df.reset_index(drop=True, inplace=True)
```
